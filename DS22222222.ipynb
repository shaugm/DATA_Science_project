There are inconsistencies and confusion in how "Dimensions" are represented. Currently, it's presented in two ways: 1- as D x W x H (Depth x Width x Height) in inches and 2- completely in inches. To resolve this, we aim to standardize all dimensions to be expressed completely in inches.

```python
df[['Depth','Width','Height']] = df['Dimensions'].str.split("x", expand=True)
df['Dimensions'].unique()
```

```python
#to remove quotes from "Dimensions" in DxWxH unit
df['Dimensions'] = df['Dimensions'].str.replace('"', '')

# Split the modified "Dimensions" column into separate columns
amount_df=df[['Depth', 'Width', 'Height']] 
amount_df = df['Dimensions'].str.split("x", expand=True)

# Check unique values in the "Dimensions" column
df['Dimensions'].unique()
```

```python
def convert_to_inches(dimensions):
    
    if dimensions is None or dimensions.strip() == '':
        return None
    
    if any(unit in dimensions for unit in ['D', 'W', 'H']):
        parts = dimensions.split(' x ')
        inches = [0, 0, 0]  # Default values
        for part in parts:
            value, unit = part[:-1], part[-1]
            if unit in ['D', 'W', 'H']:
                index = {'D': 0, 'W': 1, 'H': 2}[unit]
                inches[index] = float(value) * 0.393701
            else:
                return dimensions  # Return original dimensions if already in inches
        return "{:.2f} x {:.2f} x {:.2f} inches".format(*inches)
    else:
        return dimensions

df['Dimensions'] = df['Dimensions'].apply(convert_to_inches)

df['Dimensions'].unique()
```

we will convert D x W x H --> inches by multipling them by 0.393701
