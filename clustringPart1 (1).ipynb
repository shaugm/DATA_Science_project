## Introduction

### Dataset Description 

> **Printer dataset** The data we will collect is printer’s information in amazon, for rows each row represent individual printer details available on Amazon, while columns represent various printer’s features or characteristics as shown below: 

>**Asin**: The unique identifier for the printer
>
>**Brand**: The brand name of the printer.
>
>**Title**: Description detailing the printer's features.
>
>**Price**: The cost of the printer.
>
>**Keywords**: Relevant keywords associated with the printer.
>
>**Protection_plans**: Availability of protection plans for the printer.
>
>**Variants**: Different versions or models of the printer.
>
>**Categories**: Categorization of the printer within Amazon's system.
>
>**Rating**: The printer rating out of 5 stars.
>
>**Ratings_total**: Total number of ratings received by the printer.
>
>**Bestsellers_rank**: The rank of printer in the whole category.
>
>**Weight**: The weight of the printer in (pounds).
>
>**Dimensions**: Printer dimensions measured in length x width x height in (inches).
>
>**Bundles**: Any additional items bundled with the printer.
>
>**Frequently_bought_together**: Items frequently purchased alongside this printer.
>
>**Newer_model**: Indicates whether the printer is a newer model or not.
>
>**Connectivity_Technology**: The method used to connect the printer (Wi-Fi, USB, Wired, Wireless, etc.).
>
>**Printing_Technology**: The technology utilized for printing (Inkjet, Laser, Thermal, etc.).
>
>**Special_Feature**: Notable features that distinguish the printer.
>
>**Color**: The color of the printer device.
>
>**Printer_Output**: The type of output the printer produces (color, monochrome, multicolor, etc.).
>
>**Maximum_Print_Speed_Color**: Maximum print speed if Printer_Output is color, measured in pages per minute (ppm).
>
>**Max_Printspeed_Monochrome**: Maximum print speed if Printer_Output is monochrome, measured in pages per minute (ppm).

```python
def convert_to_inches(dimensions):
    
    if dimensions is None or dimensions.strip() == '':
        return None
    
    if any(unit in dimensions for unit in ['D', 'W', 'H']):
        parts = dimensions.split(' x ')
        inches = [0, 0, 0]  # Default values
        for part in parts:
            value, unit = part[:-1], part[-1]
            if unit in ['D', 'W', 'H']:
                index = {'D': 0, 'W': 1, 'H': 2}[unit]
                inches[index] = float(value) * 0.393701
            else:
                return dimensions  # Return original dimensions if already in inches
        return "{:.2f} x {:.2f} x {:.2f} inches".format(*inches)
    else:
        return dimensions

df['Dimensions'] = df['Dimensions'].apply(convert_to_inches)

df['Dimensions']
```

The most important argument in this clustring is n_clusters, which specifies how many clusters to place the observations in.

However, we don’t know beforehand how many clusters is optimal so we must create a plot that displays the number of clusters along with the SSE (sum of squared errors) of the model.

We will create this type of plot wehre we look for an “elbow” where the sum of squares begins to “bend” or level off. This is typically the optimal number of clusters.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

#initialize kmeans parameters
kmeans_kwargs = {
"init": "random",
"n_init": 10,
"random_state": 1,
}

#create list to hold SSE values for each k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(df)
    sse.append(kmeans.inertia_)

#visualize results
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```
